`labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']`

1. Seleccione solo las columnas *name* y *score*.
  
2. Filtre las filas donde el número de intentos (attemps) sea mayor que 2.

```python
# Definiendo el diccionario
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
             'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

# Definiendo los índices
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Construyendo el DataFrame a partir del diccionario y la lista labels
df = pd.DataFrame(exam_data, index = labels)

# Imprimiendo el DataFrame construído
df
```

```python
# Punto 1: Seleccione sólo las columnas name y score

df_filtered_1 = df.loc[:,['name','score']]
df_filtered_1
```

```python
# Punto 2: Filtre las filas donde el número de intentos (attemps) sea mayor que 2.

df_filtered_2 = df[df['attempts'] > 2]
df_filtered_2
```

```python
# Punto 3: Filtre las filas donde el score se encuentre entre 15 y 20 (incluidos)

df_filtered_3 = df[(df['score'] >= 15) & (df['score'] <= 20)]
df_filtered_3
```

```python
# Punto 4: Cambie el score en la fila 'd' a 11.5

print(f"Valor anterior en ('d','score') era: {df.loc['d','score']}")
df.loc['d','score'] = 11.5
print(f"Valor actual en ('d','score') es: {df.loc['d','score']}")
```

```python
# Punto 5: Agregue una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna
df.loc['k'] = {'name': "Juan",
               'score': 15,
               'attempts': 2,
               'qualify' : 'yes'}

df
```

```python
# Punto 6: Borre la fila 'k' y entregue nuevamente el DataFrame original
df.drop('k', inplace = True)
df
```

```python
# Punto 7: Ordene el DataFrame por nombre en orden ascendente
df.sort_values('name', ascending = True, inplace=True)
df
```

```python
# Punto 8: Reemplace los valores de la columna qualify que contiene los valores "yes" y "no" por los valores booleanos True y False respectivamente.

df['qualify'] = df['qualify'].map({'yes': True,'no':False})
df
```

## **Ejercicio 2** (30%)

Importe el dataset `athletes` del repositorio del curso y realice las siguientes tareas:

1. Importarlo como nombre athletes
2. Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.
3. Calcule la masa muscular por fila con una función como

$$I_{m.m.} = \frac{Weight}{Height^2}$$

4. Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

5. ¿Cuál es el pais con más participación femenina?

```python
# Punto 1: Importarlo como nombre athletes
athletes = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv")
athletes.info()
```

```python
athletes.head()
```

```python
# Punto 2: Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.

athletes_filtered = athletes[(athletes['gold'] > 0) | (athletes['silver'] > 0) | (athletes['bronze'] > 0)]
athletes_filtered
```

```python
# Punto 3: Calcule la masa muscular por fila

def masa_muscular(row):
  '''
  La función masa_muscular calcula el índice de masa muscular (IMM) para una fila de datos.

  Parámetros:
  row (pd.Series): Fila de datos con las columnas 'weight' y 'height'.

  Retorna:
  float: Índice de masa muscular (IMM) calculado.
  '''
  imm = row.loc['weight'] / (row.loc['height']**2)
  return imm

athletes['imm'] = athletes.apply(masa_muscular, axis =1)
athletes
```

```python
# Punto 4: Agrupe los datos calculando la media por deporte para cada una del resto de características.

# Seleccionando sólo las columnas numéricas para las cuales hace sentido calcular el promedio.
athletes_mean = athletes.loc[:,['sport','height','weight','gold','silver','bronze','imm']].groupby('sport').mean()
athletes_mean
```

```python
# Punto 4: ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?
athletes_mean.reset_index(drop= False, inplace= True)

# Encontrando el índice del deporte con el mayor peso promedio
mean_max = athletes_mean.weight.idxmax()

# Accediendo al deporte correspondiente utilizando .loc
sport_max_mean_weight = athletes_mean.loc[mean_max, 'sport']

# Respuesta
print(f"El deporte en que los jugadores tiene mayor peso promedio es: {sport_max_mean_weight}")
```

```python
# Punto 5: ¿Cuál es el país con más participación femenina?
female_participation = athletes[athletes['sex'] == 'female'].loc[:,['id','nationality']].groupby('nationality').size().idxmax()

# Respuesta
print(f"El país con más participación femenina es: {female_participation}")
```

## **Ejercicio 3 (30%)**

Importe el dataset de nba del repositorio de clase y responda (hint: Elimine las instancias con registros faltantes):

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)
2. Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información. Muestre un gráfico de barras con el número de jugadores por grupo de edad diferenciando en esta (mediante un código de color) la posición.
3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

```python
# Importar el dataset nba
nba = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv", index_col= 'Unnamed: 0')

#Cambiar nombre de columnas a minúscula
nba.columns = nba.columns.str.lower()
nba.head()

```

```python
# Verificando si hay instancias con registros faltantes
nba.info()
```

```python
# Eliminando las instancias con registros faltantes
nba.dropna(inplace= True)
nba.info()
```

```python
#1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)
df_salario = nba.groupby('team').salary.sum().sort_values(ascending= False)

# Crear un nuevo DataFrame con los resultados
df_salario = pd.DataFrame({
    'team': df_salario.index,
    'nomina': df_salario.values
})

print("DataFrame con el Ranking de los equipos según la cantidad de dinero de la nómina:")
df_salario
```

```python
#2. Separe la variable Age en tres grupos y cree una nueva columna llamada Age-Group con esta información.
nba['Age-Group'] = pd.cut(nba.age, bins= 3)
nba
```

```python
# Muestre un gráfico de barras con el número de jugadores por grupo de edad diferenciando en esta
# (mediante un código de color) la posición.
import matplotlib.pyplot as plt

# Crear la figura y el gráfico de barras
nba_figure = nba.groupby(['Age-Group', 'position']).size().unstack()
ax = nba_figure.plot(kind='bar', stacked=True)

# Configurar el gráfico
plt.title('Número de jugadores por grupo de edad')
plt.xlabel('Grupo de edad')
plt.ylabel('Número de jugadores')
plt.xticks(rotation=0);
```

```python
# Punto 3: Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para
# los cinco equipos que pagan mejores salarios.
df_salario_top5 = df_salario.loc[0:4,'team'].tolist()

nba[nba['team'].isin(df_salario_top5)].pivot_table('salary',
                                                      index = 'team',
                                                      columns = 'position',
                                                      aggfunc= 'mean').style.format(precision=2
                                                                                    , thousands="."
                                                                                    , decimal=",")
```
